```python
import pandas as pd

# Load raw CNN training codes
cnn_df = pd.read_csv("data/raw/CNN_Model_Train_Data.csv")

cnn_df.head()
```

```python
# Clean StockCode: keep only letters and digits
cnn_df['StockCode'] = (
    cnn_df['StockCode']
    .astype(str)
    .str.replace(r'[^A-Za-z0-9]', '', regex=True)
)

# Quick sanity check
cnn_df.head()
cnn_df['StockCode'].str.contains(r'[^A-Za-z0-9]', regex=True).sum()  # should be 0
```

```python
cnn_df.to_csv("data/processed/cnn_train_codes_clean.csv", index=False)
```
