```python
df = pd.read_csv("data/raw/dataset.csv", encoding="ISO-8859-1")
df.head()
```

```python
import re

def clean_stockcode(x):
    if pd.isna(x):
        return x
    return re.sub(r'[^A-Za-z0-9]', '', str(x))

df['StockCode'] = df['StockCode'].apply(clean_stockcode)
df['StockCode'].head()
```

```python
df['Quantity'] = df['Quantity'].astype(str).str.replace(r'[^0-9\-]', '', regex=True)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Quantity'].head()
```

```python
df['CustomerID'] = df['CustomerID'].astype(str).str.replace(r'[^0-9]', '', regex=True)
df['CustomerID'] = pd.to_numeric(df['CustomerID'], errors='coerce')
df['CustomerID'].head()
```

```python
df['Description'] = df['Description'].astype(str).str.replace(r'[^A-Za-z0-9 \-\'\"]', '', regex=True)
df['Description'].head()
```

```python
df['Country'] = df['Country'].astype(str).str.replace(r'[^A-Za-z ]', '', regex=True)
df['Country'].head()
```

```python
# Final clean-up for Country column
df['Country'] = df['Country'].str.replace(r'[^A-Za-z ]', '', regex=True).str.strip()

df['Country'].unique()[:10]
```

```python
df['Country'] = df['Country'].str.replace(r'^[A-Za-z]{2,4}(?=[A-Z])', '', regex=True).str.strip()
```

```python
df['Country'] = df['Country'].str.replace(r'^[A-Za-z]{2,4}(?=[A-Z])', '', regex=True).str.strip()

df['Country'].unique()
```

```python
df.head()
```

### Initial Findings from Data Exploration

- The `Country` column contained unexpected prefixes (e.g., `XxYUnited Kingdom`).
- These prefixes followed a pattern of 2–4 uppercase/lowercase letters attached directly to the country name.
- No numeric or special-character contamination was found—just unwanted alphabetic prefixes.
- A regex rule was applied to remove the prefix while keeping the valid country name intact.
- After cleaning, the `Country` column contained only standard country names (e.g., United Kingdom, France, Germany).

```python
# Convert InvoiceDate column to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

df.info()
```

```python
df[pd.to_numeric(df['UnitPrice'], errors='coerce').isna()].head(10)
```

```python
# Remove "Ww" prefix and convert to numeric
df['UnitPrice'] = df['UnitPrice'].astype(str).str.replace('Ww', '', regex=False)
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
```

```python
df[pd.to_numeric(df['UnitPrice'], errors='coerce').isna()].head()
```

```python
df.duplicated().sum()
df[df['Quantity'] < 0].head()
df[df['UnitPrice'] <= 0].head()
```

```python
# Remove invalid prices
df = df[df['UnitPrice'] > 0]

# Remove invalid quantities (keep only positive sold items)
df = df[df['Quantity'] > 0]

# Drop duplicates
df = df.drop_duplicates()

# Drop rows missing CustomerID (optional, but improves quality)
df = df.dropna(subset=['CustomerID'])

df.info()
```

```python
# Final clean-up for InvoiceNo: keep only digits
df['InvoiceNo'] = (
    df['InvoiceNo']
    .astype(str)
    .str.extract(r'(\d+)', expand=False)  # grabs only the numeric part
)
```
